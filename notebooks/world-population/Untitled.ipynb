```python
import pandas as pd
```

```python
p = pd.read_csv("2021_population.csv")
```

```python
p
```

```python
p["2021_last_updated"] = p["2021_last_updated"].str.replace(",","").astype(float)
p["2020_population"] = p["2020_population"].str.replace(",","").astype(float)
```

```python
p["area"] = p["area"].str.replace(",","").str.replace(" sq_km","").astype(float)
p["density_sq_km"] = p["density_sq_km"].str.replace(",","").str.replace("/sq_km","").astype(float)
p["growth_rate"] = p["growth_rate"].str.replace("%","").astype(float)
```

```python
p.head()
```

```python
p.to_csv("2021_population_cleaned.csv", index=False)
```
