```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
policydata=[]
countrydata=[]
yeardata=[]
statusdata=[]
jurisdictiondata=[]
for j in range(1,187):
    print(j)
    link = "https://www.iea.org/policies?page=" + str(j)
    print(link)
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    policy = soup.find_all(class_="m-policy-listing-item__link")
    countries = soup.find_all(class_="m-policy-listing-item__col m-policy-listing-item__col--country")
    year = soup.find_all(class_="m-policy-listing-item__col m-policy-listing-item__col--year")
    status = soup.find_all(class_="m-policy-listing-item__col m-policy-listing-item__col--status")
    jurisdiction = soup.find_all(class_="m-policy-listing-item__col m-policy-listing-item__col--jurisdiction")
    

    for i in range(len(policy)):
        policy_name = policy[i].get_text(strip=True)
        policydata.append(policy_name)
    for i in range(len(countries)):
        countries_name = countries[i].get_text(strip=True)
        countrydata.append(countries_name)
    for i in range(len(year)):
        year_name = year[i].get_text(strip=True)
        yeardata.append(year_name)
    for i in range(len(status)):
        status_name = status[i].get_text(strip=True)
        statusdata.append(status_name)
    for i in range(len(jurisdiction)):
        jurisdiction_name = jurisdiction[i].get_text(strip=True)
        jurisdictiondata.append(jurisdiction_name)
final = pd.DataFrame(
    {
        "Policy": policydata,
        "Country": countrydata,
        "Year": yeardata,
        "Status": statusdata,
        "Jurisdiction": jurisdictiondata,
    }
)

```
